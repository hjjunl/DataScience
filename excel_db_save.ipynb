```python
import pandas as pd
import numpy as np

from collections import deque
import pymysql
import pandas as pd
from openpyxl import Workbook
from openpyxl import load_workbook
import io
import pandas as pd
filename1 = '/content/drive/My Drive/excel/data_20y_v21.xlsx'
data = pd.read_excel(filename1)
data.head()

#DB Delete All
def delete_all():
    conn = pymysql.connect(host='hostname', user='username',db='databasename', passwd='password', charset='utf8')
    try:
        with conn.cursor() as curs:
            sql = 'delete from cc_costbill'
            curs.execute(sql)
        conn.commit()
    finally:
        conn.close()

def select_all():
    conn = pymysql.connect(host='hostname', user='username',db='databasename', passwd='password', charset='utf8')
    try:    
        with conn.cursor() as curs:
            sql = "select * from cc_costbill"
            curs.execute(sql)
            rs = curs.fetchall()
            for row in rs:
                print(row)
    finally:
        conn.close()


def insert_excel_to_db():
    conn = pymysql.connect(host='hostname', user='username',db='databasename', passwd='password', charset='utf8')
    try:
        with conn.cursor() as curs:
            sql = 'insert into cc_costbill (id,	version_cd,	periodym_cd,	item_cd,	bi_brm,	ra_rm,	ei_erm,	ei_elc,	ei_eoh,	ic_dlfc, ic_ohdfe,ic_ohdfd,	ic_dlvc,	ic_ohdvc,	ic_idlc,	ic_idohc,	proq,	proamt_unit,ic_arm,	uc_srw,	uc_dlc,	uc_idlc,	uc_idohc,	uc_dohc) values(%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s)'

            wb = load_workbook(filename1,data_only=True)
            ws = wb['Sheet1']
            
            iter_rows = iter(ws.rows)
            next(iter_rows)
            for row in iter_rows:
              curs.execute(sql, (int(row[0].value),row[1].value,int(row[2].value), row[3].value, int(row[4].value),int(row[5].value), int(row[6].value), int(row[7].value), int(row[8].value),int(row[9].value), int(row[10].value),
              int(row[11].value), int(row[12].value), int(row[13].value),int(row[14].value), int(row[15].value), int(row[16].value), int(row[17].value), 
              int(row[18].value), int(row[19].value), int(row[20].value), int(row[21].value),int(row[22].value),int(row[23].value)))             

            conn.commit()
    finally:
        conn.close()
        wb.close()
 
 
if __name__ == "__main__":
    delete_all()
    insert_excel_to_db()
    select_all()
```
